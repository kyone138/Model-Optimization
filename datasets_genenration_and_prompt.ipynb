```python
from datasets import load_dataset
dataset = load_dataset('hotpot_qa','fullwiki')
```

```python
def flatten_list_convet_to_text(whole_contex):
  answers = []
  for title,sent in zip(whole_contex['title'],whole_contex['sentences']):
    ans = "title: "+ title + "sentences: " + " ".join(sent)
    answers.append(ans)
  return " ".join(answers)
```

```python
text_list = []
for instance in dataset['train']:
  text_list.append(str("Question: " + instance['question'] + "\t" + "Context: " + flatten_list_convet_to_text(instance['context']) + "\t" + "type: " + instance['type'] + "\t" + "Answer: " + str(instance['answer'])))
```

```python
with open('train_hotpot.txt', 'w') as f:
    for line in text_list:
        f.write(f"{line}\n")
```

```python
text_list = []
for instance in dataset['validation']:
  text_list.append(str("Question: " + instance['question'] + "\t" + "Context: " + flatten_list_convet_to_text(instance['context']) + "\t" + "type: " + instance['type'] + "\t" + "Answer: " + str(instance['answer'])))
```

```python
with open('val_hotpot.txt', 'w') as f:
    for line in text_list:
        f.write(f"{line}\n")
```

```python
text_list = []
for instance in dataset['test']:
  text_list.append(str("Question: " + instance['question'] + "\t" + "Context: " + flatten_list_convet_to_text(instance['context']) + "\t" + "Answer: " ))
```

```python
with open('test_hotpot.txt', 'w') as f:
    for line in text_list:
        f.write(f"{line}\n")
```

```python
import os
```

```python
def club_txt_files(data_dir):


  train_file = os.path.join(data_dir, 'train.txt')
  val_file = os.path.join(data_dir, 'val.txt')
  test_file = os.path.join(data_dir, 'test.txt')

  with open(train_file, 'w') as f:
    for filename in os.listdir(data_dir):
      if filename.startswith('train_'):
        with open(os.path.join(data_dir, filename)) as infile:
          f.write(infile.read())

  with open(val_file, 'w') as f:
    for filename in os.listdir(data_dir):
      if filename.startswith('val_'):
        with open(os.path.join(data_dir, filename)) as infile:
          f.write(infile.read())

  with open(test_file, 'w') as f:
    for filename in os.listdir(data_dir):
      if filename.startswith('test_'):
        with open(os.path.join(data_dir, filename)) as infile:
          f.write(infile.read())

data_dir = 'data'
club_txt_files(data_dir)
```
